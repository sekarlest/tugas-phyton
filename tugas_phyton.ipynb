```python
gaji = 20000000
punya_anak = True
punya_hutang = True

if gaji > 3000000:
    print ("Gaji diatas UMR")
    if punya_anak:
        print ("Wajib beri anak uang saku")
    else:
        print ("Tidak perlu siapkan uang saku")

    if punya_hutang:
        print ("wajib bayar hutang")
    else:
        print ("tidak wajib bayar hutang")
else:
    print ("Gaji masih dibawah UMR")
```
